```python
import numpy as np
import matplotlib.pyplot as plt
from nnfs.datasets import vertical_data
```

# Dense Layer class

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.n_inputs = n_inputs
        self.n_neurons = n_neurons
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

# ReLU Activation Class

```python
class Activation_ReLU:
    def forward(self, dense_layer_outputs):
        self.output = np.maximum(0, dense_layer_outputs)
```

# Softmax Activation Function

```python
class Activation_Softmax:
    def forward(self, final_layer_outputs):
        exp_values = np.exp(final_layer_outputs - np.max(final_layer_outputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
```

# Categorical Cross Entropy Loss Function class

```python
class Loss:
    def calculate(self, y_pred, y_true):
        neg_loss_liklihoods = self.forward(y_pred, y_true)
        avg_loss = np.mean(neg_loss_liklihoods)
        return avg_loss

class Loss_CategoricalCrossEntropy(Loss):
    def forward(self, y_pred, y_true):
        y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)

        if len(y_true.shape) == 1:
            correct_confidences = y_pred_clipped[range(len(y_pred_clipped)), y_true]
        elif len(y_true.shape) == 2:
            correct_confidences = np.sum(y_pred_clipped*y_true, axis=1)

        neg_loss_liklihoods = -np.log(correct_confidences)

        return neg_loss_liklihoods
```

# Accuracy Function class

```python
class Accuracy:
    def calculateAcc(self, outputs, y_true):
        correct_perdictions = np.argmax(outputs, axis=1)
        class_targets = y_true

        if len(y_true.shape) == 2:
            class_targets = np.argmax(y_true, axis=1)

        accuracy = np.mean(correct_perdictions == class_targets)

        return accuracy
```

# Model class

```python
class Model(Loss_CategoricalCrossEntropy, Accuracy):
    def __init__(self, network_architecture):
        self.network_architecture = network_architecture

    def fit(self, training_data, y_true, epochs):
        self.best_weights = []
        self.best_biases = []
        self.dense_layers = []
        
        for layer in self.network_architecture:
            if isinstance(layer, Layer_Dense):
                self.dense_layers.append(layer)
                
        for dense_layer in self.dense_layers:
            self.best_weights.append(dense_layer.weights.copy())
            self.best_biases.append(dense_layer.biases.copy())
                
        self.lowest_loss = 999999 # initial value for lowest loss

        
        for epoch in range(epochs):
            for dense_layer in self.dense_layers:
                dense_layer.weights += 0.05 * np.random.randn(dense_layer.n_inputs, dense_layer.n_neurons)
                dense_layer.biases += 0.05 * np.random.randn(1, dense_layer.n_neurons)

            # Neural network forward pass
            progressive_input = training_data # Initial input will be the training data
            for layer in self.network_architecture:
                layer.forward(progressive_input)
                progressive_input = layer.output

            softmax_outputs = progressive_input

            self.loss = self.calculate(softmax_outputs, y_true)
            self.accuracy = self.calculateAcc(softmax_outputs, y_true)

            if self.loss < self.lowest_loss:
                print("Better weights and biases found at epoch:", epoch+1, " Loss:", self.loss, " Accuracy:", self.accuracy)
                self.new_best_weights = []
                self.new_best_biases = []
                
                for dense_layer in self.dense_layers:
                    self.new_best_weights.append(dense_layer.weights.copy())
                    self.new_best_biases.append(dense_layer.biases.copy())

                self.best_weights = self.new_best_weights
                self.best_biases = self.new_best_biases

                self.lowest_loss = self.loss
            else:
                for layer, w, b in zip(self.dense_layers, self.best_weights, self.best_biases):
                    layer.weights = w.copy()
                    layer.biases = b.copy()

        print("Completed training process!")
```

# Loading the dataset

```python
X, y = vertical_data(samples=200, classes=3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='brg')
plt.show()
```

# Architecture and Training

```python
network_architecture = [
    Layer_Dense(2, 3),
    Activation_ReLU(),
    Layer_Dense(3, 3),
    Activation_Softmax()
]

model = Model(network_architecture=network_architecture)
model.fit(training_data=X, y_true=y, epochs=100000)
```

```python
# Trying this strategy in virtical dataset
layer1 = Layer_Dense(2, 3)
activation_relu = Activation_ReLU()

layer2 = Layer_Dense(3, 3)
activation_softmax = Activation_Softmax()

loss_function = Loss_CategoricalCrossEntropy()
accuracy_function = Accuracy()

best_weights_layer1 = layer1.weights.copy()
best_biases_layer1 = layer1.biases.copy()
best_weights_layer2 = layer2.weights.copy()
best_biases_layer2 = layer2.biases.copy()

lowest_loss = 999999 # some random initial value

for i in range(100000):
    layer1.weights += 0.05 * np.random.randn(2, 3)
    layer1.biases += 0.05 * np.random.randn(1, 3)
    layer2.weights += 0.05 * np.random.randn(3, 3)
    layer2.biases += 0.05 * np.random.randn(1, 3)

    layer1.forward(X)
    activation_relu.forward(layer1.output)
    layer2.forward(activation_relu.output)
    activation_softmax.forward(layer2.output)

    loss = loss_function.calculate(activation_softmax.output, y)
    accuracy = accuracy_function.calculateAcc(activation_softmax.output, y)

    if loss < lowest_loss:
        print("New set of weights found, iteration:", i, "loss: ", loss, "accuracy: ", accuracy)
        best_weights_layer1 = layer1.weights
        best_biases_layer1 = layer1.biases
        best_weights_layer2 = layer2.weights
        best_biases_layer2 = layer2.biases
        lowest_loss = loss
    else:
        layer1.weights = best_weights_layer1.copy()
        layer1.biases = best_biases_layer1.copy()
        layer2.weights = best_weights_layer2.copy()
        layer2.biases = best_biases_layer2.copy()

print("Best pair of weights and biases found with the lowest loss")
```
