```python
import numpy as np
from sklearn.datasets import make_moons, make_blobs
import matplotlib.pyplot as plt
```

# Dense Layer class

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.n_inputs = n_inputs
        self.n_neurons = n_neurons
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = 0.01 * np.random.randn(1, n_neurons)

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) * self.biases
```

# ReLU Activation Class

```python
class Activaton_ReLU:
    def forward(self, dense_layer_outputs):
        self.output = np.maximum(0, dense_layer_outputs)
```

# Softmax Activation Function

```python
class Activation_Softmax:
    def forward(self, final_layer_outputs):
        exp_values = np.exp(final_layer_outputs - np.max(final_layer_outputs, axis=1, keepdims=True))
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
```

# Categorical Cross Entropy Loss Function class

```python
class Loss_CategoricalCrossEntropy:
    def calculate(self, neg_log_liklihoods, y_true):
        neg_log_liklihoods = self.forward(neg_log_liklihoods, y_true)
        avg_loss = np.mean(neg_log_liklihoods)
        return avg_loss
    
    def forward(self, y_pred, y_true):
        y_pred_clipped = np.clip(y_pred, 1e-7, 1-1e-7)
        correct_confidences = y_pred_clipped[range(len(y_pred_clipped)), y_true]
        neg_log_liklihoods = -np.log(correct_confidences)
        return neg_log_liklihoods
```

# Accuracy Function class

```python
class Accuracy:
    def calculateAcc(self, y_pred, y_true):
        correct_confidences_indexes = np.argmax(y_pred, axis=1)
        accuracy = np.mean(correct_confidences_indexes == y_true)
        return accuracy
```

# Model class

```python
class Model(Loss_CategoricalCrossEntropy, Accuracy):
    def __init__(self, network_architecture):
        self.network_architecture = network_architecture

    def fit(self, traning_data, y_true, epochs):
        self.best_weights = []
        self.best_biases = []
        for layer in network_architecture:
            if isinstance(layer, Layer_Dense):
                self.best_weights.append(layer.weights.copy())
                self.best_biases.append(layer.biases.copy())
                
        self.lowest_loss = 999999 # initial value for lowest loss

        self.dense_layers = []
        for layer in self.network_architecture:
            if isinstance(layer, Layer_Dense):
                self.dense_layers.append(layer)
        
        for epoch in range(epochs):
            for dense_layer in self.dense_layers:
                dense_layer.weights += 0.05 * np.random.randn(dense_layer.n_inputs, dense_layer.n_neurons)
                dense_layer.biases += 0.05 * np.random.randn(1, dense_layer.n_neurons)

            # Neural network forward pass
            progressive_input = traning_data # Initial input will be the training data
            for layer in self.network_architecture:
                layer.forward(progressive_input)
                progressive_input = layer.output

            softmax_outputs = progressive_input

            self.loss = self.calculate(softmax_outputs, y_true)
            self.accuracy = self.calculateAcc(softmax_outputs, y_true)

            if self.loss < self.lowest_loss:
                print("Better weights and biases found at epoch:", epoch+1, " Loss:", self.loss, " Accuracy:", self.accuracy)
                self.new_best_weights = []
                self.new_best_biases = []
                
                for dense_layer in self.dense_layers:
                    self.new_best_weights.append(dense_layer.weights.copy())
                    self.new_best_biases.append(dense_layer.biases.copy())

                self.best_weights = self.new_best_weights
                self.best_biases = self.new_best_biases

                self.lowest_loss = self.loss
            else:
                for i in range(len(self.dense_layers)):
                    self.dense_layers[i].weights = self.best_weights[i]
                    self.dense_layers[i].biases = self.best_biases[i]

        print("Completed training process!")
                
```

# Loading the  make moons dataset

```python
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)
print("Shape of sample:", X.shape)
print("Shape of class targets:", y.shape)
print("Samples in dataset ->")
print(X[:10])
```

```python
print("Class Targets (y_true) ->")
print(y[:10])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Make moons dataset")
plt.show()
```

# Architecture and Training

```python
network_architecture = [
    Layer_Dense(2, 4),
    Activaton_ReLU(),
    Layer_Dense(4, 2),
    Activation_Softmax()
]

model = Model(network_architecture)
model.fit(traning_data=X, y_true=y, epochs=2000)
```
