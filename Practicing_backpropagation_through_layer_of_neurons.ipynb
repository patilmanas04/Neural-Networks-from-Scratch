```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining relu activation function and relu derivative

```python
def relu(x):
    return np.maximum(0, x)

def derivative_relu(x):
    return np.where(x>0, 1, 0)
```

# Training and optimizing weights and biases using backpropagation and gradient decent

```python
inputs = np.array([1.2, -0.7, 3.1, 0.5])
weights = np.array([
    [0.2, -0.5, 1.3, 0.7],
    [-0.3, 0.8, -1.2, 0.6],
    [1.1, 0.2, -0.4, -1.5],
    [-0.9, 0.3, 0.9, 1.2],
    [0.5, -1.1, 0.8, -0.6]
])
biases = np.array([0.1, -0.2, 0.05, 0.3, -0.1])
true_value = 1.0
learning_rate = 0.001

losses = []

for epoch in range(200):
    # forward pass
    z = np.dot(weights, inputs) + biases
    a = relu(z)
    y = np.sum(a)

    loss = (y-true_value) ** 2
    losses.append(loss)

    if loss > losses[epoch-1]:
        break

    print(f"Epoch: {epoch+1} Loss: {loss}")

    dL_dy = 2*y - true_value
    dy_da = np.ones_like(a)
    da_dz = derivative_relu(a)

    dL_dz = dL_dy * dy_da * da_dz
    dL_dw = np.outer(dL_dz, inputs)
    dL_db = dL_dz

    # Updating weights and biases through gradient decent
    weights -= learning_rate * dL_dw
    biases -= learning_rate * dL_db

print(f"\nUpdated weights ->\n{weights}\n")
print(f"Updated biases ->\n{biases}")
```

# Plotting the graph for loss

```python
plt.figure(figsize=(10, 6)) # Optional: set figure size
plt.plot(range(1, len(losses)+1), losses)
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True) # Optional: add a grid
plt.show()
```
